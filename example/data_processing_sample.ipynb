```python
import pandas as pd
import numpy as np
import utils.aunz as aunz 

df_deliveroo, df_uber, df_eatnow, df_google = aunz.getstep1output()

"""kitchen type list by country""" 
list_kitchen_by_country = ['Indian',  'Chinese', 'Turkish', 'Thai',  'Nepalese', 
                             'Mexican', 'Malaysian', 'Vietnamese', 'Italian', 'British',
                             'American', 'Greek', 'Indonesian', 'Australian', 'African', 
                             'French', 'Brazilian', 'Sri Lankan', 'Lebanese','Japanese','Korean'] 

"""kitchen type list by food type""" 
food_type_lst= ['Sushi', 'Pizza','Seafood', 'Curry', 'Vegan', 'Burgers', 'Salads', 
                'Salad','Dessert','Noodles','Ramen', 'Kebab', 'Chips','Donuts','Pasta','Café', 
                'Cafe', 'Seafood']  
```

```python
%%time 
def clean_table(tb0): 
        tbl=tb0.copy()
        tbl['kitchen']= tbl.loc[:,'kitchen'].astype(str).apply(lambda x: x.split(';')) 
        tbl['row_number']=tbl.index  
        try: 
            if 'phone' in tbl.columns: 
                tbl['phone']=[tbl['phone'][i][1:] for i in range(len(list(tbl.axes[0])))] 
        except KeyError as e: 
            pringt(e)
            pass
        return tbl 
    
def country(table): 
        return [i for k in range(len(table.index)) \
                   for i in table['kitchen'][k]  \
                for j in list_kitchen_by_country if i==j]
def index_1(table):         
        return [k for k in range(len(table.index)) \
                for i in table['kitchen'][k]  for j \
                in list_kitchen_by_country if i==j] 
def keep(table):
    table= table.loc[~table.duplicated(subset= 'row_number', keep= 'last'), :] 
    return table 

def food(table):
        return [i for k in range(len(table.index)) for i in table['kitchen'][k]  \
                for j in food_type_lst if i==j]  
def index_2(table):
    return [k for k in range(len(table.index)) for i in table['kitchen'][k] \
                for j in food_type_lst if i==j]   
```

```python
def country_type(): 
    output= list()
    for d in [df_deliveroo, df_uber, df_eatnow]: 
        df= pd.DataFrame() 
        df['country_type']= country(clean_table(d)) 
        df['row_number']= index_1(clean_table(d)) 
        output.append(pd.merge(clean_table(d), keep(df), on= 'row_number', how= 'left') )
    return output  

deliveroo, ubereats, eatnow = country_type() 
```

```python
def food_type(): 
    output2= list() 
    for d in [df_deliveroo, df_uber, df_eatnow]: 
        df2= pd.DataFrame() 
        df2['food_type']= food(clean_table(d))
        df2['row_number']= index_2(clean_table(d))  
        output2.append(pd.merge(clean_table(d), keep(df2), on='row_number', how= 'left'))  
    return output2 

deliveroo_food,ubereats_food,eatnow_food= food_type() 
```

```python
df_de= pd.merge(deliveroo, deliveroo_food.loc[:,['row_number', 'food_type']], on= 'row_number', how= 'left') 
df_ub= pd.merge(ubereats, ubereats_food.loc[:,['row_number', 'food_type']], on= 'row_number', how= 'left') 
df_en= pd.merge(eatnow, eatnow_food.loc[:,['row_number', 'food_type']], on= 'row_number', how= 'left')   
df_de 
```

```python
def getstep15output():   
    df_de.to_csv('DATA/storesInfoDeliverooAU_step15.csv', index=False)  
    df_ub.to_csv('DATA/storesInfoUbereatsAUNZ_step15.csv', index= False) 
    df_en.to_csv('DATA/storesInfoEatNowAUNZ_step15.csv', index= False) 
    return df_de, df_uber, df_eatnow 

d1, d2, d3=getstep15output()  
```
